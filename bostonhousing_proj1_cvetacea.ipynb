# Описание данных таблицы

**crim** - 
уровень преступности на душу населения.

**zn** -
доля земель под жилую застройку на участках площадью более 25 000 кв. футов.

**indus** -
доля неторгового бизнеса в акрах на город.

**chas** -
фиктивная переменная реки Чарльз (= 1, если участок граничит с рекой; 0 в противном случае).

**nox** - 
концентрация оксидов азота (частей на 10 млн).

**rm** - 
среднее количество комнат в жилом доме.

**age** - 
доля жилых единиц, построенных до 1940 года.

**dis** - 
средневзвешенное расстояние до пяти бостонских центров занятости.

**rad** - 
индекс доступности к радиальным магистралям.

**tax** - 
полная ставка налога на имущество за 10 000 долларов США.

**ptratio** - 
соотношение учеников и учителей по городам.

**black** - 
1000(Bk - 0,63)^2, где Bk — доля чернокожих в городе.

**istat** - 
более низкий статус населения (в процентах).

**medv** -
средняя стоимость домов, занимаемых владельцами, в 1000 долларов.




# График зависимости параметров **dis** от **medv**.

Присвоим x_data значения в колонке **dis**, а y_data - **medv**

```python
x_data = bh_data['medv']
y_data = bh_data['dis']
plt.plot(x_data, y_data,'o')
```

# Средняя цена

Задание: создать отдельный DataFrame, содержащий только данные для домов, чья цена выше средней цены всех домов в исходных данных.
Данные о стоимости домов находятся в столбце **medv**.
Для начала посчитаем среднее значение в этом столбце.

```python
bh_data['medv'].mean()
```

Для удобства округляем полученное значение до десятых:

```python
round(22.768768768768783, 1)
```

Теперь имея среднее значение отбираем все данные о домах, чья цена выше средней и вносим их в новый DataFrame -  **rich_boston**:

```python
rich_boston = bh_data[bh_data['medv'] >= 22.8]
rich_boston.head(10)
```

Полученный DataFramr заносим в файл **csv**:

```python
rich_boston.to_csv("rich_boston.csv")
```
