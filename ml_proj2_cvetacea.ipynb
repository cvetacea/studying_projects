## Практика 2. Обучение модели.

Подгружаем библиотеки, данные:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/mcsml/ml_basics/master/hw2.csv')
data.head(10)
```

Строим график данных:

```python
plt.plot(data['X'], data['Y'], 'o', alpha=0.5)
plt.grid(True)
```

Загружаем алгоритм линейной регрессии из **scikit-learn** в модель:

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

Обучаем модель на данных:

```python
model.fit(data['X'].values.reshape(-1, 1), data['Y'].values.reshape(-1, 1))
```

Наша модель обучилась. Используем метод **predict** для предсказания данных:

```python
pred = model.predict(data['X'].values.reshape(-1, 1))
```

Строим график

```python
plt.plot(data['X'], data['Y'], 'o', alpha=0.5)
plt.plot(data['X'], pred)
plt.grid(True)
```

## Boston Housing 2

Импортируем данные

```python
bh_data = pd.read_csv('https://raw.githubusercontent.com/mcsml/ml_basics/master/train.csv')
bh_data.head()
```

Составляем уже известный нам график зависимости средней стоимости домов **medv** от расстояния до центров занятости **dis**

```python
plt.plot(bh_data['dis'], bh_data['medv'], 'go', alpha=0.5)
plt.xlabel('Расстояние до центров занятости')
plt.ylabel('Средняя стоимость домов, в 1000$')
plt.title('Зависимость средней стоимости домов от расстояния до центров занятости\n')
plt.grid(True)
```

Подгружаем модель. Мы обучим ее предсказывать стоимость домов.

```python
boston_model = LinearRegression()
boston_model
```

"Скармливаем" модели данные:

```python
boston_model.fit(bh_data['dis'].values.reshape(-1, 1), bh_data['medv'].values.reshape(-1, 1))
```

Используя метод **predict** обучаем модель предсказывать значения по исходным данным:

```python
price_pred = boston_model.predict(bh_data['dis'].values.reshape(-1, 1))
```

Строим график предсказания:

```python
plt.plot(bh_data['dis'], bh_data['medv'], 'go', alpha=0.5)
plt.plot(bh_data['dis'], price_pred)
plt.xlabel('Расстояние до центров занятости')
plt.ylabel('Средняя стоимость домов, в 1000$')
plt.title('Зависимость средней стоимости домов от расстояния до центров занятости\n')
plt.grid(True)
```
