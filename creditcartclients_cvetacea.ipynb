```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df_credit = pd.read_csv(r"C:\Users\teply\Documents\Studying\MCS\UCI_Credit_Card.csv")
```

```python
df_credit.head()
```

1. Предобработайте данные.

```python
df_credit = df_credit.rename(columns=str.lower) 
df_credit.info()
```

2. Разделите данные на тренировочный и тестовый датасеты

```python
x = df_credit.drop(['default.payment.next.month'], axis = 1) #делим данные на features X и lables Y
y = df_credit['default.payment.next.month']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
x_train #проверим, как разделилось
```

3. Создайте простейший классификатор без регуляризации и обучите его на тренировочных данных.
Создайте простейший классификатор с регуляризацией и обучите его на тех же тренировочных данных.
Сравните точность двух классификаторов.
Оцените precision и recall для каждого из результатов.

```python
from sklearn.linear_model import LogisticRegression
```

```python
lrc = LogisticRegression(max_iter = 500)
lrc.fit(x_test, y_test)
```

```python
lrc.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
print('LogisticRegression')
print('train accuracy: {}' .format(accuracy_score(y_train, lrc.predict(x_train))))
print('train precision: {}' .format(precision_score(y_train, lrc.predict(x_train))))
print('train recall: {}' .format(recall_score(y_train, lrc.predict(x_train))))
```

```python
lrc2 = LogisticRegression(max_iter = 500, C=0.1)
lrc2.fit(x_test, y_test)
```
